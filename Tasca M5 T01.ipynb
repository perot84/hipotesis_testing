```python
import pandas as pd
db= pd.read_csv ('Lligues europees UEFA.csv',sep=';', encoding = 'latin1')
db
```

Significat columnes: https://fbref.com/en/expected-goals-model-explained/

```python
from scipy.stats import shapiro

data = db['xGD']
stat, p = shapiro(data)
print('stat=%.3f, p-valor=%.3f' % (stat, p))
if p > 0.025: #en aquest cas marco 0.025 ja que hi ha valors positius i negatius
	print('Probable Gaussian')
else:
	print('Probable no Gaussian')
```

La variable té una distribució Gaussiana, per tant NO rebutja la hipòtesis nul·la

- Exercici 2

Amb el mateix dataset selecciona dos altres atributs del conjunt de dades. Calcula els p-valors i digues si rebutgen la hipòtesi nul·la agafant un alfa de 5%.

```python
from scipy.stats import pearsonr

data1 = db['xG']
data2 = db['xGA']
stat, p = pearsonr(data1, data2)
print('stat=%.3f, p-valor=%.3f' % (stat, p))
if p > 0.05:
	print('Probabement independent')
else:
	print('Probablement dependent')
```

Les variables són dependents per tant SI que rebutjem la hipòtesis nul·la, ja que aquesta determina que són valors independents

- Exercici 3

Continua amb el conjunt de dades adjunt i selecciona tres atributs del conjunt de dades. Calcula el p-valor i digues si rebutja la hipòtesi nul·la agafant un alfa de 5%.

```python
from scipy.stats import f_oneway
data1 = db['xG']
data2 = db['xGA']
data3 = db['xGD']
stat, p = f_oneway(data1, data2, data3)
print('stat=%.3f, p-valor =%.3f' % (stat, p))
if p > 0.05:
	print('Probablement la mateixa distribució')
else:
	print('Probablement diferents distribucions')
```

Les variables tenen distribucions diferents, per tant SI que rebutjem la hipòtesis nul·la, ja que aquesta determina que les variables tenen la mateixa distribució
